```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
```

## Question 2

```python
# Read the tables
items = pd.read_csv("items.csv")
customers = pd.read_csv("customers.csv")
order_item = pd.read_csv("order_item.csv")
orders = pd.read_csv("orders.csv")
```

Data Analysis - Exercise 1

```python
# Calculate the average price based on the item_price column
item_average_price = items["item_price"].mean()
print("Average item price:", item_average_price)
```

Data Analysis - Exercise 2

```python
# Create new table, with order_id and its total sum
order_total_quantity = order_item.groupby('order_id')[['quantity']].sum()

# Merge the order sum to a new table
items_with_order_quantity = pd.merge(orders, order_total_quantity, on='order_id', how='left')

# Get the id of the top paying customer
customer_id = items_with_order_quantity.groupby('customer_id')[['quantity']].sum().idxmax()
print(f"ID of the top buyer is", customer_id.iloc[0])
```
